```python
import matplotlib.pyplot as plt
import rasterio
import numpy as np
from scipy.ndimage import distance_transform_edt

import higra as hg


import sys
sys.path.append(r"D:\Source\Test")


import mxnet as mx
from skimage import measure
from shapely.geometry import Polygon
import geopandas as gpd
import tools
```

```python
# Load the three bands (extent, boundary, distance) from the prediction image
extent_path = r"D:\Source\Test\TextMxnet\data\2022\BB\Output\Fractal\Result\prediction_6982_layer2_class0.tif"
boundary_path = r"D:\Source\Test\TextMxnet\data\2022\BB\Output\Fractal\Result\prediction_6982_layer2_class1.tif"
```

```python
def InstSegm(extent, boundary, t_ext=0.4, t_bound=0.2):
    """
    INPUTS:
    extent : extent prediction
    boundary : boundary prediction
    t_ext : threshold for extent
    t_bound : threshold for boundary
    OUTPUT:
    instances
    """

    # Threshold extent mask
    ext_binary = np.uint8(extent >= t_ext)

    # Artificially create strong boundaries for
    # pixels with non-field labels
    input_hws = np.copy(boundary)
    input_hws[ext_binary == 0] = 1

    # Create the directed graph
    size = input_hws.shape[:2]
    graph = hg.get_8_adjacency_graph(size)
    edge_weights = hg.weight_graph(
        graph,
        input_hws,
        hg.WeightFunction.mean)

    tree, altitudes = hg.watershed_hierarchy_by_dynamics(
        graph,
        edge_weights)

    # Get individual fields
    # by cutting the graph using altitude
    instances = hg.labelisation_horizontal_cut_from_threshold(
        tree,
        altitudes,
        threshold=t_bound).astype(np.float)

    instances[ext_binary == 0] = np.nan

    return instances
```

```python
# Load extent and boundary masks
with rasterio.open(extent_path, 'r') as src:
    extent = src.read(1)  # Read extent mask


with rasterio.open(boundary_path, 'r') as src:
    boundary = src.read(1)  # Read boundary mask

distance = distance_transform_edt(boundary <= 0.4)
print(f"Distance Mask - Min: {distance.min()}, Max: {distance.max()}, Mean: {distance.mean()}")
```

```python
# Visualize extent, boundary, and distance
def plot_masks(extent, boundary, distance):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(extent, cmap='viridis')
    axes[0].set_title("Extent Mask")
    axes[1].imshow(boundary, cmap='viridis')
    axes[1].set_title("Boundary Mask")
    axes[2].imshow(distance, cmap='viridis')
    axes[2].set_title("Distance Mask")
    plt.tight_layout()
    plt.show()

plot_masks(extent, boundary, distance)
```

```python
# Save the distance mask as GeoTIFF (optional)
distance_path = r"D:\Source\Test\TextMxnet\data\2022\BB\Output\Fractal\Result\distance_mask.tif"
with rasterio.open(boundary_path, 'r') as src:  # Use boundary file metadata as template
    profile = src.profile
    profile.update(dtype=rasterio.float32)

    with rasterio.open(distance_path, 'w', **profile) as dst:
        dst.write(distance.astype(rasterio.float32), 1)

print(f"Distance mask saved to {distance_path}")
```

```python
# Generate instance segmentation output
t_ext = 0.5
t_bound = 0.2
out = InstSegm(extent, boundary, t_ext=t_ext, t_bound=t_bound)
```

```python
# Visualize instance segmentation
plt.figure(figsize=(10, 10))
plt.imshow(out, cmap='prism', interpolation=None)
plt.title("Instance Segmentation Output")
plt.show()
```

```python
print(out)
print(f"Data type: {type(out)}, Data shape: {out.shape}")
```

```python
# Convert the numpy array (segmentation output) to MXNet NDArray
out_mxnet = mx.nd.array(out)

# Function to process MXNet NDArray for vectorization
def vectorize_with_mxnet(array_mx, nan_value=np.nan):
    """
    Converts MXNet NDArray segmentation results into vectorized polygons.
    
    Parameters:
        array_mx (mxnet.nd.NDArray): Input segmentation array in MXNet format.
        nan_value (float): Value to ignore during processing.
        
    Returns:
        geopandas.GeoDataFrame: GeoDataFrame containing polygons.
    """
    # Convert to numpy array for processing
    array = array_mx.asnumpy()
    array = np.nan_to_num(array, nan=-1)

    polygons = []
    labels = []

    for region_label in np.unique(array):
        if region_label == -1:  # Skip placeholder (-1, representing NaN)
            continue

        # Extract contours for each labeled region
        contours = measure.find_contours(array == region_label, 0.5)

        for contour in contours:
            # Ensure the contour has at least 3 points to form a polygon
            if len(contour) < 3:
                continue

            # Convert contour to a Shapely polygon
            polygon = Polygon(contour[:, ::-1])  # Reverse coordinates to (x, y)

            # Keep valid polygons only
            if polygon.is_valid:
                polygons.append(polygon)
                labels.append(region_label)

    # Create GeoDataFrame with polygons and their labels
    gdf = gpd.GeoDataFrame({'geometry': polygons, 'label': labels}, crs="EPSG:3035")

    return gdf

# Vectorize the MXNet NDArray using the adapted function
vectorized_mxnet_gdf = vectorize_with_mxnet(out_mxnet)



# Display the GeoDataFrame to the user
tools.display_dataframe_to_user(name="Vectorized Polygons - MXNet Processed", dataframe=vectorized_mxnet_gdf)
```
